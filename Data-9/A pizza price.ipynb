```python
import numpy as np
import pandas as pd
import matplotlib as plt
from matplotlib import pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("pizza_data.csv")
```

```python
df.head(3)
```

```python
df.tail()
```

```python
df.info(verbose=True) 
```

```python
# get the number of missing data points per column
missing_values_count = df.isnull().sum()
# look at the missing points
missing_values_count[0:5]
```

```python
# let's replace that dollar sign
df["Price"] = df["Price"].str.replace("$", "").astype("float64")
# let's have a look
df.head()
# and let's get the information
df.info(verbose=True)
```

```python
df.describe(exclude=[np.number])
```

```python
# set the figure
plt.figure( figsize=(24, 9.5))
# construct countplot
sns.countplot( data=df, x="Company")
# seting x-label
plt.xlabel( xlabel="Company",  weight="bold",fontsize=225
)
# setting y-label
plt.ylabel( ylabel="Count",weight="bold",  fontsize=22.5)
# setting x-ticks
plt.xticks(weight="bold",  fontsize=15)
# setting y-ticks
plt.yticks(weight="bold", fontsize=15 );
```

```python
# figuring size
plt.figure(
    figsize=(24, 10)
)
# countplot
sns.countplot(
    data=df, 
    x="Type",
)
# x-label
plt.xlabel(
    xlabel="Type", 
    weight="bold", 
    fontsize=22.5
)
# y-label
plt.ylabel(
    ylabel="Count", 
    weight="bold", 
    fontsize=22.5
)
# x-ticks
plt.xticks(
    weight="bold", 
    fontsize=12.5
)
# y-ticks
plt.yticks(
    weight="bold", 
    fontsize=12.5
);
```
